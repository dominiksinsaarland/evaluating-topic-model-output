```python
#!/usr/bin/env python3

import pandas as pd

df_rating_scores = pd.read_json("../computed/ratings_outfile_with_logit_bias_old_prompt_5_runs.jsonl", lines=True)
df_grouped = df_rating_scores.groupby(["dataset_name", "model_name", "topic_id"])["chatGPT_eval", "human_eval"].mean().reset_index()
df_grouped["diff_abs"] = (df_grouped["chatGPT_eval"]-df_grouped["human_eval"]).abs()
df_grouped["diff"] = (df_grouped["chatGPT_eval"]-df_grouped["human_eval"])
```

```python
df_grouped.sort_values(by="diff_abs")[:5]
```

```python
df_grouped.sort_values(by="diff")[-5:]
```
